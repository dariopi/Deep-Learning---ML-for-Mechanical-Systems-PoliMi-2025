```python
# solution

class FeedforwardNN(nn.Module):
    def __init__(self, input_dim, hidden1, hidden2):
        super(FeedforwardNN, self).__init__()
        self.fc1 = nn.Linear(input_dim, hidden1)
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Linear(hidden1, hidden2)
        self.relu2 = nn.ReLU()
        self.fc3 = nn.Linear(hidden2, 1)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu1(x)
        x = self.fc2(x)
        x = self.relu2(x)
        y = self.fc3(x)
        return y


model = FeedforwardNN(input_dim = X_train.shape[1], hidden1=12, hidden2=8)

print(f"Model structure: {model}")

for name, params in model.named_parameters():
    print(f"parameter name: {name}. Value {params.data}")

# check what model provides:
y_hat = model(X_train)
y_hat.shape
```

```python
# Training loop with accuracy calculation (solution)
max_epochs = 20000
for it in range(max_epochs):
    optimizer.zero_grad()
    y_hat = model(X_train)  # Predictions (probabilities with Softmax)
    loss = criterion(y_hat, y_train)  # Compute loss
    loss.backward()
    optimizer.step()

    if it % 5 == 0:  # Print every 5 iterations
        print(f"Iteration: {it}. Loss: {loss.item():.3f}")

    
```

## Step 6. Assess model performance


```python
# Assess final results (solution)
def assess_results(model, X, y, data_type, scaler_y):
    with torch.no_grad():
        y_hat = model(X)

        # Convert tensors to numpy for denormalization
        y_hat_numpy = y_hat.numpy()
        y_numpy = y.numpy()

        # Denormalize the output
        y_hat_denorm = scaler_y.inverse_transform(y_hat_numpy)
        y_denorm = scaler_y.inverse_transform(y_numpy)

        # Prediction and plotting
        plt.figure(figsize=(12, 12))
        plt.plot(y_hat_denorm, 'r-', label='Predicted output')
        plt.plot(y_denorm, 'b-', label='True output')
        plt.title(f'Predicted vs Estimated output: {data_type} dataset')  
        plt.legend()
        plt.show()

        # Scatter plot
        plt.figure(figsize=(5, 5))
        plt.scatter(y_hat_denorm, y_denorm)
        plt.xlabel('Predicted output')
        plt.ylabel('True output')
        plt.title(f'Predicted vs Estimated output: {data_type} dataset')  
        plt.legend()
        plt.show()
        
        # RMSE calculation
        mse = torch.mean((y_hat - y) ** 2)
        rmse = torch.sqrt(mse)

        # R2 calculation
        total_variance = torch.var(y)
        residual_variance = torch.mean((y_hat - y) ** 2)
        R2 = 1 - residual_variance / total_variance

    return rmse, R2 
```

```python
# Assess results in training
print('Training results')

rmse, R2 = assess_results(model, X_train, y_train, 'train', scaler_y)
print(f"rmse = {rmse}. R2 = {R2:.3f}")


# Assess results in test
print('Test results')

rmse, R2 = assess_results(model, X_test, y_test, 'test', scaler_y)
print(f"rmse = {rmse}. R2 = {R2:.3f}")
```
